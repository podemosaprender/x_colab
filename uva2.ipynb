##**¿Qué son las UVI, las UVA y dólar blue ?**

  * El dólar blue es la divisa que se compra y se vende en el mercado informal practicamente en forma inmediata.  

  * UVI,unidades de cuenta para invertir, ahorrar o tomar préstamos. Son unidades de cuenta porque están relacionadas con el precio de otros bienes y servicios: 1000 UVIs valen lo mismo que un metro cuadrado construido.

  *UVA, es la Unidad de Valor Adquisitivo actualizable por "CER", valor que se actualiza diariamente y lo podés consultar en la página del Banco Central de la República Argentina.



 ###**¿Qué es un plazo fijo y que significa que sea precancelable?**

Se trata de un depósito de dinero que podés hacer por un tiempo determinado. Este mismo puede ser precancelable o no, esto significa que el cliente puede cancelarlo antes del tiempo pactado inicialmente con el banco. En la mayoría de los casos es [precancelable](https://www.bcra.gob.ar/bcrayvos/Preg-Frec-Se-puede-cancelar-anticipadamente-un-Plazo-Fijo.asp) al cumplirse los 30 días de corrido. La posibilidad de cancelarlo antes de tiempo suele tardar unos días en algunos casos hasta 5 días hábiles.

# Datos comparables, merge tables.

```python
#TODO: convertir todos en INDICES que valgan lo mismo en el primer dia ej. 100 y cambien proporcionalmente.
indices_df= pd.merge(uva_df, uvi_df, on="Fecha")
indices_df.rename(columns={'Valor_x':'UVA', 'Valor_y': 'UVI','indice_x':'UVA_index', 'indice_y':'UVI_index'}, inplace=True)

indices_df
```

```python
#TODO: convertir todos en INDICES que valgan lo mismo en el primer dia ej. 100 y cambien proporcionalmente.
indices_df= pd.merge(indices_df, usd_blue_df, on="Fecha")
indices_df.rename(columns={'indice':'i_USD'}, inplace=True)
indices_df
```

###¿Cuantos fideos compra el dólar blue? 

Quiero contar que, si compraste dólares y aumento menos que los fideos, ahora te alcanza para menos fideos.


```python
indices_df['i_fideos']=  indices_df['i_USD'] / indices_df['UVA_index'] 
indices_df['i_carry']= indices_df['UVA_index']  / indices_df['i_USD'] 
```

```python
# de la tabla indices_df me traigo las columnas 'Fecha', 'i_fideos'
fig = px.line(
    indices_df,
    x="Fecha",
    y='i_fideos',
    title='Poder compra dólar.'
  )
fig.show()
```

```python
f = px.line(
    indices_df,
    x="Fecha",
    y=["i_carry"],
    title='¿Cuantos dólares compro si me quedo en UVA?'
  )
f.show()
```

# TODO: Datos entre dos fechas 

```python
Fecha1='2022-02-04'# 
Fecha2='2022-07-16'
Fecha3='2022-09-01'

def datos_entre(Fecha1,Fecha2):
  indices_1_2 = indices_df[(indices_df['Fecha']> Fecha1) & (indices_df['Fecha']< Fecha2)].copy()
  indices_1_2['UVA_index'] = indices_1_2['UVA_index'] / (indices_1_2.iloc[0]['UVA_index']) * 100
  indices_1_2['UVI_index'] = indices_1_2['UVI_index'] / (indices_1_2.iloc[0]['UVI_index']) * 100
  indices_1_2['i_USD'] = indices_1_2['i_USD'] / (indices_1_2.iloc[0]['i_USD']) * 100
  return indices_1_2

def grafico_entre(Fecha1,Fecha2):
  df=datos_entre(Fecha1,Fecha2)
  fig = px.line(df, x="Fecha", y=["UVA_index","UVI_index","i_USD"], title='Indice de precios Construcción (UVI) vs Consumidor (UVA) vs Dolar Blue')
  fig.show()
    
```

# **¿Qué más podemos aprender del gráfico?**
  De los gráficos a continuación podemos concluir que, en el primero, se observa que la mejor inversión es la del plazo fijo UVA por un periodo bastante prolongado están muy por encima de las demás. En el segundo solo por un pequeño periodo el dólar blue se disparó, pero luego volvió a sus valores en menos de 15 días.

Muestra que no necesariamente por un incremento en la cotización o en rendimiento de una inversión debemos cambiarnos. Solo deberíamos cambiar de cuando tenemos algún tipo de certeza en que se puede llegar a tener una corriente alcista y esta puede ser estable.
